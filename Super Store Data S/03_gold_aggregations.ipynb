```python
from pyspark.sql.functions import sum

df = spark.table("workspace.sales.silver_superstore")

display(df)
```

```python
gold_df = df.groupBy("Category").agg(
    sum("Sales").alias("total_sales"),
    sum("Profit").alias("total_profit")
)

display(gold_df)
```

```python
spark.sql("create schema if not exists workspace.gold")
from pyspark.sql import functions as F
gold_df.write.mode("overwrite").saveAsTable("workspace.gold.sales_by_category")

display(spark.sql("select * from workspace.gold.sales_by_category"))
```

```python
from pyspark.sql.functions import year, month, sum

df = spark.table("workspace.sales.silver_superstore")

monthly = df.groupBy(
    year("Order_Date_Clean").alias("year"),
    month("Order_Date_Clean").alias("month")
).agg(
    sum("Sales").alias("total_sales")
)

monthly = monthly.orderBy("year", "month")

monthly.write.mode("overwrite").saveAsTable("workspace.gold.monthly_sales")
display(spark.sql("select * from workspace.gold.monthly_sales"))
```
